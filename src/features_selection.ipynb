```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df1=pd.read_csv('input_data.csv')
```

```python
df2=pd.read_csv('inventory_data.csv')
df2['inventory_index']=df2.iloc[:,0]
```

```python
df2.head()
```

```python
#calulating Correlation of "inventory" with other features. 
corr_inventory=df1.corrwith(df2['Inventory'],method='pearson').abs()
print(corr_inventory)
```

```python
#Selecting highly correlated features
relevant_features = corr_inventory[corr_inventory>0.5]
```

```python
print(relevant_features.sort_values(ascending=False).head(20))
```

```python
selected_df = df1[relevant_features.index]  # Only getting columns having corr> 0.5 wrt output
selected_df['input_index']=df1.iloc[:,0]
```

```python
selected_df.describe()
```

```python
df_all=pd.merge(
    df2,
    selected_df,
    left_on=['inventory_index'],
    right_on=['input_index']
)
```

```python
df_all.head()
```

```python
df_all=df_all.drop(columns=['Inventory (t-1)','inventory_index'])
```

```python
df_all.head()
```

```python
pearsoncorr = selected_df.corr(method='pearson').abs()
```

```python
columns = np.full((pearsoncorr.shape[0],), True, dtype=bool)
for i in range(pearsoncorr.shape[0]):
    for j in range(i+1, pearsoncorr.shape[0]):
        if pearsoncorr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
```

```python
filtered_df= selected_df.columns[columns]
```
